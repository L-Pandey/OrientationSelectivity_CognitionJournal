```python
# SELECT AND LOAD MODEL
def init_model(model, model_path=None):
    if model == 'vit':
        model = LitClassifier.load_from_checkpoint(model_path).backbone
        model.fc = nn.Identity()
    elif model == 'untrained_vit':
        configuration = ViTConfigExtended()
        configuration.num_hidden_layers = 3
        configuration.num_attention_heads = 3
        # print configuration parameters of ViT
        print('image_size - ', configuration.image_size)
        print('patch_size - ', configuration.patch_size)
        print('num_classes - ', configuration.num_classes)
        print('hidden_size - ', configuration.hidden_size)
        print('intermediate_size - ', configuration.intermediate_size)
        print('num_hidden_layers - ', configuration.num_hidden_layers)
        print('num_attention_heads - ', configuration.num_attention_heads)
        
        # pass the configuration parameters to get backbone
        backbone = Backbone('vit', configuration)
        model = LitClassifier(backbone).backbone
        model.fc = nn.Identity()
    elif model == 'simclr':
        model = SimCLR.load_from_checkpoint(model_path)

    model.eval()
    return model
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
from sklearn.neighbors import NeighborhoodComponentsAnalysis
from torchvision.transforms import ToPILImage


def plot_embedding_kittenAI(model, data_loader, title, method='tsne', plot_points=True, plot_images=False):
    features = []
    labels = []
    imgs = [] # extra

    # Extract features.
    model.eval()
    torch.manual_seed(0)
    with torch.no_grad():
        for i, (img, label) in enumerate(data_loader):
            if i > 25:
                break
            output = model(img)
            features.append(output.cpu().numpy())
            labels.append(label.cpu().numpy())
            imgs.append(img) # extra

    X = np.concatenate(features, axis=0)
    y = np.concatenate(labels)
    z = np.concatenate(imgs) # extra
    
    print("len of x - ", len(X))
    print("len of y - ", len(y))
    print("len of z - ", len(z))
    # Choose dimensionality reduction method.
    if method == 'PCA':
        dim_reduction =  PCA(n_components=2)
    elif method == 'tSNE':
        dim_reduction = TSNE(n_components=2, verbose=1, perplexity=25, n_iter=4000)


    results = dim_reduction.fit_transform(X)
    fig, ax = plt.subplots()
    fig.set_dpi(100)
    
    
    if plot_points:
        #cmap_1 = matplotlib.cm.get_cmap('winter')
        #cmap_2 = matplotlib.cm.get_cmap('autumn')
        # iterate over the number of classes
        for i in range(10): # 0째-90째
            label = int(dataset.classes[i])
            
            if label == 0:
                color = cmap_self8(label/10)
            elif label == 10:
                color = cmap_self(label/10)
            elif label == 20:
                color = cmap_self7(label/10)
            elif label == 30:
                color = cmap_self9(label/10)
            elif label == 40:
                color = cmap_self4(label/10)
            elif label == 50:
                color = cmap_self2(label/10)
            elif label == 60:
                color = cmap_self12(label/10)
            elif label == 70:
                color = cmap_self11(label/10)
            elif label == 80:
                color = cmap_self10(label/10)
            elif label == 90:
                color = cmap_self3(label/10)
            #color = cmap_self3(label / 10) if label < 50 else cmap_self6(label / 10)
            ax.scatter(
                results[y==i, 0], results[y==i, 1], # this is similar to results[:, 0], results[:, 1], except we don't get different clusters then
                label=str(label)+'째', # added in original
                alpha=0.5,
                color=color,
            )
    #ax.legend()
    #ax.set_title(method)
    if title:
        ax.set_title(title, fontsize=60)
    ax.set_xticks([])
    ax.set_yticks([])
    #ax.legend(loc='upper left', shadow=True, scatterpoints=1, bbox_to_anchor=(1.05, 1))

    # my code - 
    # Plot images.
    if plot_images:
        to_pil_image = ToPILImage()

        for x0, y0, im in zip(results[:,0], results[:,1], z):
            img_temp = torch.from_numpy(im)
            #print(img_temp.shape)
            #img = img_temp.permute(2,1,0)
            #print(img.shape)
            img = to_pil_image(img_temp)
            img.thumbnail((24,24))
            ab = AnnotationBbox(OffsetImage(img), (x0, y0), frameon=False)
            ax.add_artist(ab)
```

```python
# Steps -
# 1. Study the code and thoroughly understand what's happening in the code
# 2. Just Run the code for a few layers, see if it works for orientation selectivity, use colored tSNEs
# to make sure the script is correct, the code should work for the last layer.
# 3. If it looks like it works, then create a class with well-written code.
# 4. Do the same for Shape-Based vision project and see what layers develop this.
# 5. If orientation selectivity is shown early in layers even a little bit, we can say that we found evidence of some selectivity in the early layers.

import torch
import torchvision.transforms as T
from PIL import Image

def get_intermediate_outputs(model, image_paths, layer_names):
    """Extracts outputs from specified layers of a frozen PyTorch model."""

    model.eval()
    intermediate_outputs = {name: [] for name in layer_names}

    device = next(model.parameters()).device  # Get device of model's parameters

    for image_path in image_paths:
        img = Image.open(image_path).convert("RGB")
        transform = T.Compose([  # Your transformations here (MUST MATCH TRAINING)
            T.Resize((64,64)), #Example
            T.ToTensor(),
            #T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) #Example
        ])
        img_tensor = transform(img).unsqueeze(0).to(device)  # Transform and move to device

        handles = []
        for name in layer_names:
            try:
                layer = model
                for part in name.split('.'):
                    layer = getattr(layer, part)

                def hook_fn(module, input, output, layer_name=name): # Define hook function inside the loop
                    intermediate_outputs[layer_name].append(output.detach().cpu())

                handle = layer.register_forward_hook(hook_fn)
                handles.append(handle)
            except AttributeError:
                print(f"Layer '{name}' not found in the model. Check the name.")
                return {}
            except Exception as e:
                print(f"Error getting layer '{name}': {e}")
                return {}


        with torch.no_grad():
            model(img_tensor)

        for handle in handles:
            handle.remove()

    return intermediate_outputs



device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Define device *FIRST*

model = model.to(device)  # Move the model to the device *immediately* after loading
model.eval()  # Set to eval mode


image_paths = ["/data/lpandey/KittenAI_Dataset/LinearProbeTrainTest/test/0/"]
layer_names = ["model.model.transformer.layers.0.0"] # Example layer names or indices

intermediate_outputs = get_intermediate_outputs(model, image_paths, layer_names)

print(intermediate_outputs.keys())

if intermediate_outputs:
    for layer_name, outputs in intermediate_outputs.items():
        print(f"Outputs from layer '{layer_name}':")
        for output in outputs:
            print(output.shape)
            # ... process the output ...
else:
    print("Failed to get intermediate outputs.")
```
