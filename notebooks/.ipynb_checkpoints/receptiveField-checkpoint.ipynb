# Setup Class: Initialize model, weights, and test directory

```python
class Setup:
    def __init__(
        self,
        model_name,
        model_path,
        heads,
        patch_size,
        image_res,
        test_dir,
    ):

        self.model_name = model_name
        self.model_path = model_path
        self.heads = heads
        self.patch_size = patch_size
        self.image_res = image_res
        self.test_dir = test_dir

        # load model w/wo weights
        self.model = self.init_model(self.model_name, 
                                     self.model_path, 
                                     self.patch_size, 
                                     self.image_res, 
                                     self.heads)

        # load frames to test the model
        self.batch = self.load_batch(self.test_dir) 
                                               


    def init_model(self, model, model_path, patch_size, image_res, heads):
        if model == 'vit':
            model = LitClassifier.load_from_checkpoint(model_path).backbone
            model.fc = nn.Identity()
            print("[INFO] ViT model with weights loaded")
            
        elif model == 'untrained_vit':
            configuration = ViTConfigExtended()
            configuration.num_attention_heads = configuration.num_hidden_layers = heads
            configuration.patch_size = 8
            configuration.image_size = 64
            # print configuration parameters of ViT
            print("[INFO] Untrained ViT model loaded")
            print('[INFO] Image Resolution - ', configuration.image_size)
            print('[INFO] Patch Size - ', configuration.patch_size)
            print('[INFO] Hidden Layers - ', configuration.num_hidden_layers)
            print('[INFO] Heads - ', configuration.num_attention_heads)
            
            # pass the configuration parameters to get backbone
            backbone = Backbone('vit', configuration)
            model = LitClassifier(backbone).backbone
            model.fc = nn.Identity()
            
        model.eval()
        return model


    def load_batch(self, test_dir, batch_size=1000, shuffle=False):
        
        image_paths = [os.path.join(test_dir, fname)
                   for fname in os.listdir(test_dir)
                   if fname.endswith(('.png', '.jpg', '.jpeg'))]

        transform = transforms.Compose([
            transforms.Resize((self.image_res, self.image_res)),
            transforms.ToTensor(),
        ])
    
        # Simple inline dataset (no custom class)
        dataset = torch.stack([(transform(Image.open(p))) for p in image_paths])
        return dataset
```

```python
##### Metric 1
import numpy as np

def receptive_field_spatial_variance(rf_map): # rf_map size or central region size here is 2
    """
    rf_map: 2D numpy array of shape [H, W], normalized gradient map
    returns: variance of the receptive field (scalar)
    """
    H, W = rf_map.shape
    x = np.arange(W)
    y = np.arange(H)
    xv, yv = np.meshgrid(x, y)

    # Normalize RF map to sum to 1 for weighted mean/var
    rf_map = rf_map / (np.sum(rf_map) + 1e-8)

    # Compute center of mass
    x_mean = np.sum(xv * rf_map)
    y_mean = np.sum(yv * rf_map)

    # Compute spatial variance
    var = np.sum(((xv - x_mean)**2 + (yv - y_mean)**2) * rf_map)
    return var


##### Metric 2
def central_activation_ratio(rf_map, center_ratio=0.5):
    """
    rf_map: 2D numpy array [H, W]
    center_ratio: defines how large the central region is (e.g., 0.5 = center half)
    """
    H, W = rf_map.shape
    h_margin = int(H * (1 - center_ratio) / 2)
    w_margin = int(W * (1 - center_ratio) / 2)

    center = rf_map[h_margin:H-h_margin, w_margin:W-w_margin]
    ratio = np.sum(center) / (np.sum(rf_map) + 1e-8)
    return ratio
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5, 6]
y1 = [0.76, 0.69, 0.5, 0.37, 0.39, 0.35]
y2 = [0.73, 0.61, 0.42, 0.42, 0.41, 0.37]
y3 = [0.79, 0.70, 0.60, 0.53, 0.46, 0.44]
y4 = [0.79, 0.73, 0.68, 0.621, 0.61, 0.53]
y5 = [0.78, 0.72, 0.70, 0.63, 0.59, 0.56]
y6 = [0.83, 0.78, 0.74, 0.71, 0.70, 0.68]

# Create the plot
plt.figure(figsize=(6, 4))

# trained line
plt.plot(x, y1, marker='o', color='blue', linestyle='-', label="160k")
plt.plot(x, y2, marker='o', color='green', linestyle='-', label="80k")
plt.plot(x, y3, marker='o', color='red', linestyle='-', label="40k")
plt.plot(x, y4, marker='o', color='orange', linestyle='-', label="20k")
plt.plot(x, y5, marker='o', color='brown', linestyle='-', label="10k")

# untrained line
plt.plot(x, y6, marker='o', color='black', linestyle='--', label="untrained")

plt.xlabel("Layers", fontsize=12)
plt.ylabel("Central Activation Ratio", fontsize=12)
plt.legend()  # This displays the label
plt.grid(True)

# Show the plot
plt.show()
```
