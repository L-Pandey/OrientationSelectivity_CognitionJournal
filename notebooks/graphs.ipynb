## Box plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your seed-wise accuracy data
data = {
    "0k": [10.0, 10.0, 10.0],
    "1k": [23.5, 21.8, 34.0],
    "2.5k": [43.0, 20.7, 40.0],
    "5k": [41.6, 44.0, 34.1],
    "10k": [91.8, 90.5, 92.0],
    "40k": [92.7, 95.0, 97.2],
    "80k": [99.6, 99.5, 99.4],
}

# Convert to long-format DataFrame for seaborn
rows = []
for train_size, acc_list in data.items():
    for acc in acc_list:
        rows.append({"Training Size": train_size, "Accuracy": acc})
df = pd.DataFrame(rows)

# Sort training sizes in logical order (custom if needed)
order = ["0k", "1k", "2.5k", "5k", "10k", "40k", "80k"]

# Plot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=df, x="Training Size", y="Accuracy", order=order, showfliers=True, width=0.6)

# Customize
#plt.title("Test Accuracy Across Training Sizes (3 Seeds Each)")
#plt.ylabel("Test Accuracy (%)")
#plt.xlabel("Training Size")
# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Graphs for Experiment 5

```python
import matplotlib.pyplot as plt
import numpy as np

# x-axis categories
x = ['L1','L2', 'L3']
x_numeric = np.arange(len(x))  # numeric x-axis for plotting

# Data for multiple subjects (each a line)
means = [
    [98.8, 93.1, 96.6],  # Ferret
    [99.3, 84.8, 93.1],  # Mouse
    [99.5, 99.6, 97.5],  # Rabbit
]

std_errors = [
    [0.06, 0.78, 0.55],       # Ferret
    [0.03, 3.8, 0.45],     # Mouse
    [0.08, 0.13, 0.18],       # Rabbit
]

colors = ['red', 'blue', 'green']
labels = ['Ferret Retinal Waves', 'Mouse Retinal Waves', 'Rabbit Retinal Waves']

# Plot
fig, ax = plt.subplots(figsize=[7, 5])

for i in range(len(means)):
    mean = np.array(means[i])
    error = np.array(std_errors[i])
    ax.plot(x_numeric, mean, '-o', label=labels[i], color=colors[i])
    ax.fill_between(x_numeric, mean - error, mean + error, color=colors[i], alpha=0.1)

# Axis settings
ax.set_ylim([0, 100])
ax.set_yticks(np.linspace(0, 100, 5))
ax.set_xticks(x_numeric)
ax.set_xticklabels(x)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show legend
ax.legend(loc='lower right', fontsize=12)

# Show plot
plt.show()
```
